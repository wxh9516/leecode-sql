```python
St=100
K=100
r=0.06
div=0.03
sigma=0.5
t=0
T=1

from scipy.stats import norm

def BSM_delta(S0,t):
    '''the sensitivity of the value of the option to a change in the underlying stock price''' 
    d1 = (np.log(S0/K)+(r-div+sigma**2/2)*(T))/(sigma*np.sqrt(T))
    delta = norm.cdf(d1)
    return delta


BSM_delta(St,t)
```

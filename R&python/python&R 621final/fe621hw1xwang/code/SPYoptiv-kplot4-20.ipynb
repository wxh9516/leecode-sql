```python
from bs4 import BeautifulSoup as soup
```

```python
from urllib.request import urlopen as uReq
```

```python
my_url = 'https://finance.yahoo.com/quote/SPY/options?p=SPY&date=1524182400'
```

```python
uClient = uReq(my_url)
```

```python
page_html = uClient.read()
```

```python
uClient.close()
```

```python
page_soup = soup(page_html,'lxml')
```

```python
page_soup.body.span
```

```python
containers=page_soup.findAll('tbody')
```

```python
calls=containers[0]
```

```python
puts=containers[1]
```

```python
len(containers)
```

```python
filename='SPYcalloptk-iv4-20.csv'
f=open(filename,'w')
```

```python
headers='contract_name,last_trade_date,strike_price,last_price,bid_price,ask_price,implied_volatility\n'
```

```python
f.write(headers)
```

```python
for container in containers:
    for call in calls:
        
        title_container=call.findAll('a',{'class':'Fz(s) Ell C($c-fuji-blue-1-b)'})
        contract_name=title_container[0].text
    
        time=call.findAll('td',{'class':'data-col1 Ta(end) Pstart(7px)'})
        last_trade_date=time[0].text
    
        strike_price2=call.findAll('td',{'class':'data-col2 Ta(end) Pstart(7px)'})
        strike_price=strike_price2[0].text
    
        lp=call.findAll('td',{'class':'data-col3 Ta(end) Pstart(7px)'})
        last_price=lp[0].text
    
        bd=call.findAll('td',{'class':'data-col4 Ta(end) Pstart(7px)'})
        bid_price=bd[0].text
    
        askp=call.findAll('td',{'class':'data-col5 Ta(end) Pstart(7px)'})
        ask_price=askp[0].text
        
        iv=call.findAll('td',{'class':'data-col10 Ta(end) Pstart(7px) Pend(6px)'})
        iv2=iv[0].text
    
        print('contract name:'+contract_name)
        print('last_trade_date:'+last_trade_date)
        print('strike price:'+strike_price)
        print('last price:'+last_price)
        print('bid price:'+bid_price)
        print('ask price:'+ask_price)
        print('implied volatility:'+iv2)
    
        f.write(contract_name+','+last_trade_date+','+strike_price.replace(',','')+','+last_price+','+bid_price+','+ask_price+','+iv2+'\n')

f.close()
```

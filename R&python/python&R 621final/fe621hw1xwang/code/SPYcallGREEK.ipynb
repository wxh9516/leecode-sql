```python
import math
import numpy as np   #manipulate mutidimensional arrays
import pandas as pd   #provide data structures
import scipy.stats as scs
import statsmodels.api as sm
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'serif' 
from scipy.integrate import quad
import mpl_toolkits.mplot3d.axes3d as p3
from scipy.stats import norm
```

```python
file1='SPYcalloptk-iv4-20.csv'
data_1=pd.read_csv(file1)
data_1
```

```python
S0=263.93
r=0.0149
T=0.25
t=0

stp=np.array(data_1['strike_price'])
stpli=stp.tolist()

iv_float = data_1['implied_volatility'].str.strip("%").astype(float)/100
```

```python

ivli=iv_float.tolist()
```

```python
filename='SPYcalloptGREEK.csv'
f=open(filename,'w')

headers='delta\n'
f.write(headers)

for i,j in zip(ivli,stpli):
    K=j
    sigma=i
    d1=(np.log(S0/K)+(r+sigma**2/2)*T)/(sigma*np.sqrt(T))
    delta = norm.cdf(d1)
    print(delta)
    
    f.write(str(delta)+'\n')

f.close()
```

```python
filename='SPYcalloptGREEKvega.csv'
f=open(filename,'w')

headers='vega\n'
f.write(headers)

for i,j in zip(ivli,stpli):
    K=j
    sigma=i
    d1=(np.log(S0/K)+(r+sigma**2/2)*T)/(sigma*np.sqrt(T))
    vega = S0 * norm.cdf(d1) * math.sqrt(T - t)
    print(vega)

    f.write(str(vega)+'\n')

f.close()
```

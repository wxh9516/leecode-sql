```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
file1='AMZNcalloptk-iv3-23.csv'
file2='AMZNcalloptk-iv4-20.csv'
file3='AMZNcalloptk-iv6-15.csv'
```

```python
data_1=pd.read_csv(file1)
data_1
```

```python
data_2=pd.read_csv(file2)
data_2
```

```python
data_3=pd.read_csv(file3)
data_3
```

```python
iv1_float = data_1['implied_volatility'].str.strip("%").astype(float)/100
iv2_float = data_2['implied_volatility'].str.strip("%").astype(float)/100
iv3_float = data_3['implied_volatility'].str.strip("%").astype(float)/100
```

```python
iv1=np.array(iv1_float)
iv_1=iv1.tolist()
sp1=np.array(data_1['strike_price'])
sp_1=sp1.tolist()

```

```python
iv2=np.array(iv2_float)
iv_2=iv2.tolist()
sp2=np.array(data_2['strike_price'])
sp_2=sp2.tolist()
```

```python
iv3=np.array(iv3_float)
iv_3=iv3.tolist()
sp3=np.array(data_2['strike_price'])
sp_3=sp3.tolist()
```

```python
plt.figure(figsize=(15, 8))
points = 158

klist = np.linspace(1000, 2100, points)
vlist = [i for i in iv_1]
plt.plot(klist, vlist)
plt.grid()
plt.xlabel('strike $K$')
plt.ylabel('implied_volatility')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
points = 228

klist = np.linspace(600, 2100, points)
vlist = [i for i in iv_2]
plt.plot(klist, vlist)
plt.grid()
plt.xlabel('strike $K$')
plt.ylabel('implied_volatility')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
points = 226

klist = np.linspace(400, 2200, points)
vlist = [i for i in iv_3]
plt.plot(klist, vlist)
plt.grid()
plt.xlabel('strike $K$')
plt.ylabel('implied_volatility')
plt.show()
```

```python
y1=iv_1
y2=iv_2
y3=iv_3
x1=range(0,len(y1))
x2=range(0,len(y2))
x3=range(0,len(y3))
plt.figure(figsize=(15, 6))
plt.plot(x1,y1,'',label="AMZNcall3-23")
plt.plot(x2,y2,'',label="AMZNcall4-20")
plt.plot(x3,y3,'',label="AMZNcall6-15")
plt.title('Implied Volatility VS Strike Price')
plt.legend(loc='upper right')
plt.xticks((0,50,100,150,200),('500','875','1250','1625','2000'))
plt.xlabel('Strike Price')
plt.ylabel('IV')
plt.grid(x1)
plt.show()
```

```python
from bs4 import BeautifulSoup as soup
```

```python
from urllib.request import urlopen as uReq
```

```python
my_url ='https://finance.yahoo.com/quote/%5EVIX/options?p=%5EVIX&date=1524009600'
```

```python
uClient = uReq(my_url)
```

```python
page_html = uClient.read()
```

```python
uClient.close()
```

```python
page_soup = soup(page_html,'lxml')
```

```python
page_soup.body.span
```

```python
containers=page_soup.findAll('tbody')
```

```python
calls=containers[0]
```

```python
puts=containers[1]
```

```python
len(containers)
```

```python
filename='VIXcallopt.csv'
f=open(filename,'w')
```

```python
headers='contract_name,last_trade_date,strike_price,last_price,bid_price,ask_price\n'
```

```python
f.write(headers)
```

```python
for container in containers:
    for call in calls:
        title_container=call.findAll('a',{'class':'Fz(s) Ell C($c-fuji-blue-1-b)'})
        contract_name=title_container[0].text
    
        time=call.findAll('td',{'class':'data-col1 Ta(end) Pstart(7px)'})
        last_trade_date=time[0].text
    
        strike_price2=call.findAll('a',{'href':'/quote/^VIX/options?strike=false&straddle=false'})
        strike_price=strike_price2[0].text
    
        lp=call.findAll('td',{'class':'data-col3 Ta(end) Pstart(7px)'})
        last_price=lp[0].text
    
        bd=call.findAll('td',{'class':'data-col4 Ta(end) Pstart(7px)'})
        bid_price=bd[0].text
    
        askp=call.findAll('td',{'class':'data-col5 Ta(end) Pstart(7px)'})
        ask_price=askp[0].text
    
        print('contract name:'+contract_name)
        print('last_trade_date:'+last_trade_date)
        print('strike price:'+strike_price)
        print('last price:'+last_price)
        print('bid price:'+bid_price)
        print('ask price:'+ask_price)
    
        f.write(contract_name+','+last_trade_date+','+strike_price.replace(',','')+','+last_price+','+bid_price+','+ask_price+'\n')

f.close()
```

```python
import pandas as pd
import math
import random
```

```python
def student_t(nu): # nu equals number of degrees of freedom
    x = random.gauss(0.0, 1.0)
    y = 2.0*random.gammavariate(0.5*nu, 2.0)
    return x / (math.sqrt(y/nu))
```

```python
file1='fe621final 3-6-last.csv'
data1=pd.read_csv(file1)
```
